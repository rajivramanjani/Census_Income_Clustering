```python
import pandas as pd
```

```python
#Read the data
cluster_data = pd.read_csv('C:/DataScience/Clustering/censusdata.csv')
```

```python
from sklearn.cluster import KMeans 
kmeans = KMeans(n_clusters=4)
```

```python
#Drop all columns that are labels and which have no value to us or which have a lot of blanks or empty values
cluster_data = cluster_data.drop([' native-country', ' capital-gain', ' capital-loss', ' income', ' workclass', ' fnlwgt'], axis=1)
cluster_data
```

```python
#Normalize data
from sklearn import preprocessing
import numpy as np
#Select only numeric data for normalization
df_num = cluster_data.select_dtypes(include=[np.number])
x = df_num.values #returns a numpy array
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x,)
df = pd.DataFrame(x_scaled)
```

```python
#Normalized values
df
```

```python
#Rename the columns
df.rename(columns={0: 'age', 1: 'education-num', 2: 'hours-per-week'}, inplace=True)
```

```python
#Dataframe after renaming
df
```

```python
df = df.reset_index()
df
```

```python
cluster_data = cluster_data.reset_index()
cluster_data
```

```python
df3 = pd.merge(df, cluster_data, left_index=True, right_index=True)
df3
```

```python
df3 = df3.drop([' education-num', 'index_x', 'index_y', 'age_y', ' hours-per-week'], axis=1)
df3.rename(columns={'age_x': 'age', ' education': 'education', ' marital-status': 'maritalstatus', }, inplace=True)
df3.rename(columns={' occupation': 'occupation', ' relationship': 'relationship', ' race': 'race', ' sex': 'sex' }, inplace=True)
df3
```

```python
# Convert string to categorical variables
df3["education"] = pd.factorize(df3.education)[0]
df3["maritalstatus"] = pd.factorize(df3.maritalstatus)[0]
df3["occupation"] = pd.factorize(df3.occupation)[0]
df3["relationship"] = pd.factorize(df3.relationship)[0]
df3["race"] = pd.factorize(df3.race)[0]
df3["sex"] = pd.factorize(df3.sex)[0]
```

```python
df3
```

```python
pd.get_dummies(df3)
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=4)
kmeans.fit(df3)
y_kmeans = kmeans.predict(df3)
```

```python
kmeans.cluster_centers_
```

```python
df3['assignments'] = kmeans.labels_
```
